#### **Import des packages**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

#### **Données d'entraînement**
Dataset cohérent pour éviter les erreurs d'Extrapolation 

```python
donnee_x = np.array([1,2,5,8,9,10,12,15,18,20,25,30,40,50,60,75,80,85,90,95,100,105,110,112,115,118,120]).reshape(-1,1)
donnee_y = np.array([76,87,110,130,135,140,155,170,175,175,175,175,175,172,170,168,167,166,165,164,162,160,158,159,158,157,156])
# print(donnee_x)
# print(donnee_y)
```

#### **Visualisation des données**

```python
sns.scatterplot(x=donnee_x.flatten(), y=donnee_y, s=80, label="Données")
```

#### **Affichage de la ligne de régression**

```python
plt.plot(donnee_x, donnee_y, label="Ligne de Régression")
```

#### **Création et entraînement du model**

```python
model = LinearRegression()
model.fit(donnee_x, donnee_y)
```

#### **Affichage des paramètres du modèle**

```python
print(f"La pente: a = {model.coef_[0]}")
print(f"Le biais: b = {model.intercept_}")
```

#### **Visualisation du modèle**

```python
sns.regplot(x=donnee_x, y=donnee_y)
plt.title('Ligne de régession du modèle')
plt.show()
```

#### **Prédictions**
